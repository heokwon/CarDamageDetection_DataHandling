#### cv2로 저장하고 불러오면 차원이 다시 생김
#### PIL 라이브러리를 사용하여 이미지를 불러온 뒤, GrayScale로 변환 후 저장

```python
import os
import PIL
from PIL import Image
from tqdm.notebook import tqdm

path = '/content/data/train/segmentation/' # GrayScale 해줄 파일경로 지정 , 파일경로 지정해줄 때 맨 뒤에 '/' 꼭 붙여주기
                                           # 폴더 '내부의' 파일들이기 때문에 경로 끝에 '/' 추가
path_list = os.listdir(path) # os.listdir(path) : path 안의 파일들을 리스트로 불러올 수 있다.
for i in tqdm(path_list): # path_list를 범위로 지정
    dir = path + i  # 경로 내에서 파일들을 하나씩 가져오기 위해 dir지정
    im_gray = Image.open(dir).convert('L') # dir에서 이미지를 불러와 convert('L')을 사용하여 GrayScale 해줌
    im_gray.save(dir[:-4]+'.png','png') # 그 뒤 '.png'로 확장자 변환후 저장해줌
    # 슬라이싱을 해준 이유는 기존 확장자 '.jpg'가 파일이름으로 들어가 있어서 그거 제거해주려고
```

```python
# 1차원으로 저장이 잘 됬는지 모든파일 size출력 후 확인해봄
for i in path_list:
  dir = path+i
  im_gray = Image.open(dir)
  print(im_gray.size)
```
