

```
seed method는 어떤 세트의 랜덤한 숫자를 생성해두고

seed가 선언된 다음에 발생하는 rand method에서 미리 정해진 세트의 난수를 return하도록 합니다.

seed method를 사용하면 몇번을 실행해도 동일한 난수를 return하도록 할 수 있습니다.
```
