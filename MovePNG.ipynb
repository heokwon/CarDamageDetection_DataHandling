```python
import os
import shutil
from tqdm.notebook import tqdm

path = '/content/multi_mask_data/images/' # 파일경로 설정
path_list = os.listdir(path) # 범위, os.listdir을 사용하여 폴더경로 내 파일들을 리스트로 불러올 수 있음

# 전체적인 해석 : path_list안에 '.png' 이름을 가진 파일들을 불러 images폴더로 이동
for i in tqdm(path_list): # path_list에 i(파일)
  if '.png' in i: # 만약 i에 '.png'가 있다면
      filename = i # i를 filename변수에 저장
      shutil.move(path+i,'/content/images/'+i) # shutil.move를 사용하여 path안의 i 들을, /content/images/에 i를 집어넣음 
```
