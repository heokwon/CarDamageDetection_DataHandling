#### .JPG -> .PNG

```python
import os
import PIL
from PIL import Image
from tqdm.notebook import tqdm

jpg_path = '/content/dataset2/val/segmentation/' # .jpg파일 경로 설정
path_list = os.listdir(jpg_path) # 폴더 안의 파일들을 리스트로 불러올수있음
new_path = '/content/dataset/val/segmentation/' # .png파일로 변환 후 저장시킬 경로 설정

for i in tqdm(path_list):
  dir = jpg_path + i # .jpg파일 경로에서 파일들을 하나씩 불러오게하기위해 jpg_path+i
  new = Image.open(dir).convert('RGB') # PIL 라이브러리의 Image를 사용하여 경로안의 이미지들을 불러온 뒤, RGB형식으로 변환시켜주고 new 변수에 저장
  new.save(new_path+i[:-4]+'.png','png') # new_path 경로에 new를 '.png'확장자로 변환하여 저장 
  # 중간의 슬라이싱 >> 저장하는 과정에서 .jpg가 파일의 이름으로 들어가있었음 -> 슬라이싱으로 제거
```
